```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import random
import pickle
import os
```

```python
datadir= 'COVID-19_Radiography_Dataset'
CATEGORIES = ['COVID', 'NORMAL', 'Viral Pneumonia']
data =[]
img_size = 100
```

```python
for category in CATEGORIES:
    folder = os.path.join(datadir,category)
    label = CATEGORIES.index(category)
    for img in os.listdir(folder):
        img_path = os.path.join(folder,img)
        img_array = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        new_array=cv2.resize(img_array, (img_size, img_size))
        data.append([new_array,label])
```

```python
random.shuffle(data)
```

```python
data_train = data[:-500]
data_test = data[-500:]
```

```python
x_train=[]
y_train=[]
for features, label in data_train:
    x_train.append(features)
    y_train.append(label)
```

```python
x_test=[]
y_test=[]
for features, label in data_test:
    x_test.append(features)
    y_test.append(label)
```

```python
pickle.dump(x_train, open('feature_gray_train.pkl','wb'))
pickle.dump(y_train, open('label_gray_train.pkl','wb'))

pickle.dump(x_test, open('feature_gray_test.pkl','wb'))
pickle.dump(y_test, open('label_gray_test.pkl','wb'))
```
