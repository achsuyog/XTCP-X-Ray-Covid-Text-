```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import random
import pickle
import os
```

```python
from tensorflow import keras
model = keras.models.load_model('C:/Users/Asus/Documents/suyog/7th sem/ML mini/mlminimodel')
```

```python
datadir= 'Prediction Data'
data =[]
img_size = 100
```

```python
for img in os.listdir(datadir):
        img_path = os.path.join(datadir,img)
        img_array = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        new_array=cv2.resize(img_array, (img_size, img_size))
        data.append([new_array])
```

```python
data = np.array(data)
data.shape
```

```python
 data = data.reshape(-1,100,100,1)
```

```python
data = data/255
```

```python
data.shape
```

```python
predictions = model.predict(data)
```

```python
new_predicted_value = []
for i in range(3490):
    predicted = np.argmax(predictions[i])
    new_predicted_value.append(predicted)
```

```python
new_predicted_value
```

```python
(unique, counts) = np.unique(new_predicted_value, return_counts=True)
frequencies = np.asarray((unique, counts)).T
frequencies
```
